```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
```

```python
cfg_D = pd.read_csv('./config_D_external_radiomic_features_20210630.csv')
cfg_E = pd.read_csv('./config_E_external_radiomic_features_20210630.csv')
```

```python
# cfg_D = pd.read_csv('./config_D_radiomic_features_20210528.csv')
# cfg_E = pd.read_csv('./config_E_radiomic_features_20210528.csv')
```

# Cfg D

```python
cols_1 = [col for col in cfg_D.columns[2:] if ('2_' not in col)]
cols_2 = [col for col in cfg_D.columns[2:] if ('2_' in col)]
```

```python
sorted(list(set([col.split('_')[0] for col in cfg_D.columns])))
```

```python
# *h와 *h1 합치기!

cfg_D_p = cfg_D[[col for col in cols_1 if col.startswith('p_')]]
cfg_D_ph = cfg_D[[col for col in cols_1 if col.startswith('ph_')]]
cfg_D_ph1 = cfg_D[[col for col in cols_1 if col.startswith('ph1_')]]
cfg_D_ph_merge = pd.DataFrame(np.where(np.isnan(cfg_D_ph1.values), cfg_D_ph.values, cfg_D_ph1.values), columns=cfg_D_ph.columns)

cfg_D_n = cfg_D[[col for col in cols_1 if col.startswith('n_')]]
cfg_D_nh = cfg_D[[col for col in cols_1 if col.startswith('nh_')]]
cfg_D_nh1 = cfg_D[[col for col in cols_1 if col.startswith('nh1_')]]
cfg_D_nh_merge = pd.DataFrame(np.where(np.isnan(cfg_D_nh1.values), cfg_D_nh.values, cfg_D_nh1.values), columns=cfg_D_nh.columns)

cfg_D_c = cfg_D[['ID', 'Contrast']+[col for col in cols_1 if col.startswith('c_')]]
cfg_D_ch = cfg_D[[col for col in cols_1 if col.startswith('ch_')]]
cfg_D_ch1 = cfg_D[[col for col in cols_1 if col.startswith('ch1_')]]
cfg_D_ch_merge = pd.DataFrame(np.where(np.isnan(cfg_D_ch1.values), cfg_D_ch.values, cfg_D_ch1.values), columns=cfg_D_ch.columns)
```

```python
cfg_D_1 = pd.concat([cfg_D_c, cfg_D_ch_merge, cfg_D_n, cfg_D_nh_merge, cfg_D_p, cfg_D_ph_merge], axis=1)
```

```python
cfg_D_1.shape
```

```python
# *2를 새로운 row로 만들기!

cfg_D_2 = cfg_D[['ID', 'Contrast']+[col for col in cols_2]]
cfg_D_2['ID'] = cfg_D_2['ID'].map(lambda x : str(x)+'_2')

cfg_D_2.columns = [col.replace('2_', '_') for col in cfg_D_2.columns]
cfg_D_2 = cfg_D_2[cfg_D_2.iloc[:,2:].isna().sum(1) < (len(cfg_D_2.columns) - 2)]
```

```python
cfg_D_2.shape
```

```python
cfg_D_total = pd.concat([cfg_D_1, cfg_D_2])
```

```python
cfg_D_total.tail(10)
```

```python
cfg_D_total['ID'] = cfg_D_total['ID'].astype(str)
cfg_D_total = cfg_D_total.sort_index(by='ID')
```

```python
sorted(list(set([col.split('_')[0] for col in cfg_D_total.columns])))
```

```python
# cfg_D_total.to_csv('./config_D_radiomic_features_20210509.csv', index=False)
cfg_D_total.to_csv('./config_D_external_radiomic_features_20210630.csv', index=False)
```
